```python
bins = [0,100000,300000,float('inf')]
labels = ['Low','Med','High']
df['Income Band'] = pd.cut(df['Estimated Income'],bins=bins,labels=labels,right=False)
```

```python
#univariate analysis
# univariate analysis means when u are analysing every columns at a time 
# we are having the data and we are anlysing the column by column
# #here im examine the distribution of unique categories in categorical columns
for i,predictor in enumerate(df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]].columns):
  plt.figure(i)
  sns.countplot(data=df,x=predictor,color ="pink")
```

```python
#bivariate analysis
# #here im examine the distribution of unique categories in categorical columns
for i,predictor in enumerate(df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]].columns):
  plt.figure(i)
  sns.countplot(data=df,x=predictor,hue = 'GenderId')
```

```python
#creating histogram to understand the distribution of data
for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title('Histogram of Occupation Count ')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
# Here we are doing the Numerical analysis
numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']
#univariate Analysis and visualization
plt.figure(figsize= (10,6))
for i , col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
    plt.show()
```

```python
#heatmaps
numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']
correaltion_matrix = df[numerical_cols].corr()
plt.figure(figsize=(6,6))
sns.heatmap(correaltion_matrix,annot=True,cmap='crest',fmt = ".2f")
plt.title("Correlation Matrix")
plt.show()
```
